# Veamos que es el clustering

Dentro de los algoritmos no supervisados, podemos ver el de clasificaci칩n o clustering.

El objetivo del clustering es separar una poblaci칩n en grupos asegurando que cada uno de ellos tenga la misma varianza. Genera conjuntos disjuntos, asignando cada punto al centroide m치s cercano. 

```python
data = pd.read_csv('heart.csv')
newdata = data[["RestingBP", "Cholesterol"]].dropna()
newdata.head()
```

```python
# N칰mero de clusters = 4, a침ado una semilla con random_state y utilizo todos los procesadores del port치til.
# Ajusto el modelo con fit.
# random_state = 0 nos permite ejecutar el modelo varias veces con la misma divisi칩n aleatoria de los datos.
kmeans = KMeans(n_clusters = 4, random_state = 0).fit(newdata)
kmeans.labels_
# Obtenemos el cluster al que pertenecen cada una de nuestras observaciones.
```

random_state = 0 nos ayuda a verificar y validar los datos cuando se ejecuta el c칩digo varias veces. Establecer random_state en un valor fijo garantizar치 que se genere la misma secuencia de n칰meros aleatorios cada vez que se ejecute el c칩digo.
fit, permite ajustar el modelo.

```python
# Con la siguiente instrucci칩n analizamos c칩mo han quedado distribuidos los individuos en cada uno de los grupos.
np.unique(kmeans.labels_, return_counts = True)
# Como se puede observar han quedado distribuidos de manera dispar
```

```python
# Vamos a visualizar c칩mo quedan distribuidos los centroides.
plt.scatter(newdata["RestingBP"], newdata["Cholesterol"], c= kmeans.labels_)
plt.xlabel('RestingBP')
plt.ylabel('Cholesterol')
plt.show()

# Valor maximo de cada columna del dataframe
newdata.max()
```

```python
# Vamos a ver d칩nde quedan situados los centroides.
kmeans.cluster_centers_
```

```python
# Vamos a mostrar c칩mo quedan clasificados los primeros 100
prediccion = kmeans.predict(newdata)[0:100]
prediccion
```

```python
# Buscar치 el n칰mero 칩ptimo de cl칰steres bas치ndose en el m칠todo del codo como se indica a continuaci칩n:
SS_distance = {}
print("Lower values are better and zero is optimal")
optimal = float("inf") # Asignamos el valor mas grande que se puede para asegurar que por grande que sea el primer valor se va a asignar a la variable 
optimal_clusters = 0

for k in range(1,20):
    mod = KMeans(n_clusters= k, max_iter=1000,init = 'k-means++')
    mod = mod.fit(newdata)
    SS_distance[k] = mod.inertia_
    # Comprobacion de cual es el mejor numero de clusters
    if SS_distance[k]<optimal:
        optimal = SS_distance[k]
        optimal_clusters = k
    # Imprimimos el resultado   
    print(str(k) +" clusters:") 
    print(SS_distance[k])
    print()
    
print("El valor mas optimo es: " + str(optimal) + " (" + str(optimal_clusters) + " clusters)")
```

```python
# Vamos a visualizarlo
plt.figure(figsize = (20,7))
sns.pointplot(x = list(SS_distance.keys()), y = list(SS_distance.values()))
plt.xlabel("N칰mero clusters")
plt.ylabel("Suma de distancias al cuadrado")
plt.title("T칠cnica del codo para encontrar el tama침o 칩ptimo de las agrupaciones")
```

Podemos observar que obtenemos tambien que el numero de clusters mas efectivo es 19

## Regresion lineal

El objetivo de este modelo es aproximar a una recta el conjunto de datos. (error del modelo).
- Es necesario que la relaci칩n entre las variables sea lineal.
- Las variables explicativas no tienen relaci칩n entre ellas.
- Homocedasticidad (varianza constante, si predecimos los salarios en base a la edad, habr칤a que encontrar los mismos errores para gente joven que para gente mayor).
- El error total sea la suma de todos los errores.

```python
data = pd.read_csv('heart.csv')
data.head()
```

```python
# Vamos a determinar las pulsaciones maximas (MaxHR) a partir del colestrol.
# Ambas variables est치n correlacionadas.
data2 = data.dropna(subset = ["MaxHR"]) # A determinar
# Seleccionamos un subconjunto de 100.000 l칤neas que vamos a mezclar.
data2 = data.sample(frac = 1).head(250)

Y = data2["MaxHR"]
X = data2[["Age"]]
```

```python
# A continuaci칩n ajustamos el modelo.
regr = linear_model.LinearRegression()
# Ajuste con X e Y definidos con anterioridad.
regr.fit(X, Y)

# Imprimimos el valor de las Betas. Corresponden con los coeficientes de las variables dependientes 
print("Coeficientes: ", regr.coef_)

# El valor pr칩ximo a 1 nos indica que est치 cercano a la varianza, desviaci칩n al cuadrado respecto a la media.

Y_pred = regr.predict(X)
# R2 nos va a decir cu치n bueno es el modelo, cuanto m치s pr칩ximo a 1, el modelo es mejor.
print("R cuadrado: ", r2_score(Y, Y_pred))
```

```python
# Mostramos los resultados de forma gr치fica, junto con la recta de regresi칩n.
plt.scatter(X[1:10000], Y[1:10000], color = "red")
plt.plot(X[1:10000], Y_pred[1:10000], color = "black")
plt.show()
```
